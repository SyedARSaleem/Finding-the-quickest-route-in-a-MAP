```python
# To display the resulting path 
def display_path(path):
    print(path[0], end = ' ')
    for i in range(1,len(path)):
        print('  --> ', path[i], end = ' ')



# Depth First Search

def dfs(graph, start, end):
    visited = []
    stack = [[start, 0]]
    shortest_distance = 10000

    # using stack in depth first search to explore the childs firstand then the neighbours
    while stack:
        current_path = stack.pop()
        node = current_path[-2]
        path_distance = current_path.pop(-1)

        if node not in visited:
            neighbours = graph[node]
            for neighbour in neighbours:
                new_path = list(current_path)
                new_path.append(neighbour[0])
                distance = path_distance + neighbour[1]
                new_path.append(distance)
                stack.append(new_path)

                if (neighbour[0] == end) and (distance < shortest_distance):
                    shortest_distance = distance
                    shortest_path = new_path

            visited.append(node)

    return shortest_path, shortest_distance


# Breadth First Search
# visit, add to queue, explore

def bfs(graph, start, end):
    visited = []
    queue = [[start, 0]]
    shortest_distance = 10000

    # usning queue to explore the level first
    while queue:
        current_path = queue.pop(0)
        node = current_path[-2]
        path_distance = current_path.pop(-1)

        if node not in visited:
            visited.append(node)
            neighbours = graph[node]
            for neighbour in neighbours:
                new_path = list(current_path)
                new_path.append(neighbour[0])
                distance = path_distance + neighbour[1]
                new_path.append(distance)
                queue.append(new_path)

                if (neighbour[0] == end) and (distance < shortest_distance):
                    shortest_distance = distance
                    shortest_path = new_path

            

    return shortest_path, shortest_distance


# A* Search

# function to calculate the heuristic distance 
def calc_heuristic(node, goal):
    # graphical coordinates roughly measured
    vector = {
        'inverness': (57.48, -4.22),
        'oban': (56.41, -5.47),
        'aberdeen': (57.15, -2.10),
        'glasgow': (55.86, -4.25),
        'edinburgh': (55.95, -3.19),
        'carlisle': (54.89, -2.92),
        'newcastle': (54.98, -1.62),
        'york': (53.96, -1.08),
        'manchester': (53.48, -2.24),
        'liverpool': (53.41, -2.99),
        'holyhead': (53.31, -4.63)
    }
    
    # calculating the heuristic distance from the given node to the goal node
    start_coord = vector[node]
    goal_coord = vector[goal]
    return ((start_coord[0] - goal_coord[0])**2 + (start_coord[1] - goal_coord[1])**2) ** 0.5    


# haepq to use priority queue for A*
import heapq

def a_star(graph, start, goal):
    visited = set()
    
    queue = [(0, [start], 0)] 
    
    while queue:
        # retrieving the values from the queue with shortest distance 
        nearest, path, total_dist = heapq.heappop(queue)
        node = path[-1]
        
        if node == goal:
            return path, total_dist
        
        if node not in visited:
            visited.add(node)
            for child in graph[node]:
                new_nearest = nearest + child[1]
                new_total_dist = total_dist + child[1]
                new_path = path + [child[0]]
                heapq.heappush(queue, (new_nearest + calc_heuristic(child[0], goal), new_path, new_total_dist))
    
    # return in case nothing found
    return None, 0


# Graph represented in a dictionary
Graph_nodes = {
    'inverness': [('oban', 110), ('aberdeen', 110), ('glasgow', 170)],
    'oban': [('inverness', 110), ('glasgow', 90)],
    'aberdeen': [('inverness', 110), ('glasgow', 140)],
    'glasgow': [('edinburgh', 40), ('carlisle', 100), ('oban', 90), ('aberdeen', 140), ('inverness', 170)],
    'edinburgh': [('manchester', 220), ('newcastle', 110), ('glasgow', 40)],
    'carlisle': [('glasgow', 100), ('manchester', 120)],
    'newcastle': [('edinburgh', 110), ('york', 80), ('manchester', 130)],
    'york': [('manchester', 60), ('newcastle', 80)],
    'manchester': [('liverpool', 40), ('york', 60), ('newcastle', 130), ('edinburgh', 220), ('carlisle', 120)],
    'liverpool': [('holyhead', 90), ('manchester', 40)],
    'holyhead': [('liverpool', 90)]
}


print("===================  Shortest Path Finder  =============================\n\n")
print("List of countries\n")
[print(i) for i in Graph_nodes.keys()]
print("\n\nHello I am going to find the shortest path between two cities for you\n")

    
start = input("\nEnter the city you want to start your journey in: ")
start = start.lower()
while start not in Graph_nodes.keys():
    print("\nWrong value Enter again\n")
    start = input("Enter the city you want to start your journey in: ")
    start = start.lower()
    
    
end = input("Enter the city you want to go to: ")
end = end.lower()
while end not in Graph_nodes.keys():
    print("\nWrong value Enter again\n")
    end = input("\nEnter the city you want to go to: ")
    end = end.lower()



print("\n\n ========================= Breadth First Search ==============================")
path, distance = bfs(Graph_nodes, start, end)
print('\n\nThe shortest path: ')
display_path(path)
print('\n\nDistance covered: ', distance)

print("\n\n ========================== Depth First Search ===============================")
path, distance = dfs(Graph_nodes, start, end)
print('\n\nThe shortest path: ')
display_path(path)
print('\n\nDistance covered: ', distance)

print("\n\n =============================== A* Search ===================================")
path, distance = a_star(Graph_nodes, start, end)
print('\n\nThe shortest path: ')
display_path(path)
print('\n\nDistance covered: ', distance)
```
